```python
reason_column = pd.get_dummies(new_data['Reason for Absence'],dtype = int)
reason_column
```

```python
new_data = new_data.drop(['Reason for Absence'], axis = 1)
```

## Grouping the reason type

```python
reason_type1 = reason_column.loc[:, 1:14].max(axis = 1)
reason_type2 = reason_column.loc[:, 15:17].max(axis = 1)
reason_type3 = reason_column.loc[:, 18:21].max(axis = 1)
reason_type4 = reason_column.loc[:, 22:].max(axis = 1)
```

```python
new_data = pd.concat([new_data, reason_type1, reason_type2, reason_type3, reason_type4], axis = 1)
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
new_data.columns = column_names
new_data.head(5)
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 
                          'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
new_data = new_data[column_names_reordered]
```

## Extraction month  value:

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
list_months = []
```

```python
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
df_reason_mod['Month Value'] = list_months
df_reason_mod.head(10)
```

## Extraction day of the week:

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod.head()
```

## Preprocessed checkpoint 

```python
data_preprocessed = df_reason_mod.copy()
data_preprocessed
```

## Creating the targets:

```python
data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours']> data_preprocessed['Absenteeism Time in Hours'].median(),1,0)
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
data_preprocessed.head()
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours','Date'], axis = 1)
data_with_targets.head()
```

## Selecting the inputs:

```python
data_with_targets.shape
```

```python
unscaled_inputs = data_with_targets.iloc[:,:-1]
unscaled_inputs.head()
```

## Standardizing the data:

```python
from sklearn.preprocessing import StandardScaler
absenteeism_scaler = StandardScaler()
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs)
scaled_inputs
```

## Spliting the data into train_test_split:

```python
from sklearn.model_selection import train_test_split
```

```python
#targets = targets.flatten()
```

```python
x_train,x_test,y_train,y_test = train_test_split(scaled_inputs,targets, test_size = 0.2, random_state = 20)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

## Logistic Regression with SKlearn:

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

## Training the model:

```python
reg = LogisticRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)*100
```

## Finding intercept and coeff:

```python
reg.intercept_
```

```python
reg.coef_
```

```python
unscaled_inputs.columns.values
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns = ['Feature Name'], data = feature_name)
```

```python
summary_table['Coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index+1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

## Interpreting the coefficients:

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
summary_table.sort_values('Odds_ratio', ascending = False)
summary_table
```

```python
reg.score(x_test,y_test)*100
```
